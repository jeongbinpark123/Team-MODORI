```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# 데이터 로드

data = pd.read_excel('달서구 MCLP 데이터프레임.xlsx')

# 히스토그램과 Q-Q 플롯을 위한 열 목록
columns_to_check = ['수요 지수', '인프라 지수', '접근성 지수', '환경 지수']

# 데이터 시각화: 히스토그램 및 Q-Q 플롯
plt.figure(figsize=(12, len(columns_to_check) * 4))

for i, column in enumerate(columns_to_check):
    plt.subplot(len(columns_to_check), 2, 2*i + 1)
    plt.hist(data[column], bins=20, density=True, alpha=0.6, color='g')
    plt.title(f'{column} 히스토그램')

    plt.subplot(len(columns_to_check), 2, 2*i + 2)
    stats.probplot(data[column], dist="norm", plot=plt)
    plt.title(f'{column} Q-Q Plot')

plt.tight_layout()
plt.show()

# 정규성 검정 (Shapiro-Wilk 테스트) 수행
normality_test_results = {}

for column in columns_to_check:
    stat, p_value = stats.shapiro(data[column])
    normality_test_results[column] = {'Statistic': stat, 'p-value': p_value}

normality_test_results
```
