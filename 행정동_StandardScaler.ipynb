```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
from matplotlib import rc
from sklearn.preprocessing import StandardScaler
import matplotlib.font_manager as fm
from sklearn.cluster import KMeans
import seaborn as sns
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
df = pd.read_excel('북구 MCLP 데이터프레임.xlsx')
df.head()
```

```python
columns_to_drop = ['행정구역', '위도', '경도']

# 열을 제거한 새로운 데이터프레임 생성
dfstandard = df.drop(columns=columns_to_drop)

# 결과 확인
dfstandard.head()
```

```python
# 숫자 데이터만 선택
df_numeric = dfstandard.select_dtypes(include='number')

# StandardScaler를 사용하여 숫자 데이터만 표준화
scaler = StandardScaler()
df_standardized_values = scaler.fit_transform(df_numeric)

# 표준화된 값을 가지고 새로운 데이터프레임 생성
df_standardized = pd.DataFrame(df_standardized_values, columns=df_numeric.columns)

# 결과 표시
df_standardized.head()
```

```python
df_standardized['행정구역'] = df['행정구역']
df_standardized['위도'] = df['위도']
df_standardized['경도'] = df['경도']
df_standardized.to_excel('df_standardized_북구.xlsx', index=False)
df_standardized = df_standardized[['행정구역', '위도', '경도', '수요 지수', '인프라 지수', '접근성 지수', '환경 지수']]
df_standardized.head()
```
