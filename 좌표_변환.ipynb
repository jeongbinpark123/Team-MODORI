```python
address = df_temp['주소']
print(address)
```

```python
for i in range(len(address)):
    a = address[i].split(' ')
    address[i] = " ".join(a[0:4])
```

```python
df_temp['주소'] = address
df_temp.head()
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
latitude = []
longitude =[]

for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
address_df = pd.DataFrame({'번호': df_temp['번호'],'소속지회':df_temp['소속지회'],'경로당명' : df_temp['경로당명'],'주소':address,'위도':latitude,'경도':longitude})

#df저장
address_df.to_csv('대구 경로당 좌표.csv')
```
